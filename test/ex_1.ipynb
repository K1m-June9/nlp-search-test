```python
# article 요약해보기
import torch
from typing import Any
from transformers import PreTrainedTokenizerFast
from transformers import BartForConditionalGeneration
```

```python
def make_summary(article: str) -> str:
    text = article
    text = text.replace('\n',' ')
    tokenizer = PreTrainedTokenizerFast.from_pretrained('digit82/kobart-summarization')
    model = BartForConditionalGeneration.from_pretrained('digit82/kobart-summarization')
    
    raw_input_ids = tokenizer.encode(text)
    input_ids = [tokenizer.bos_token_id] + raw_input_ids + [tokenizer.eos_token_id]

    summary_ids = model.generate(torch.tensor([input_ids]), num_beams=4, max_length=1024, eos_token_id = 1)
    simple = tokenizer.decode(summary_ids.squeeze().tolist(), skip_special_tokens=True)

    return simple    
```

```python
import torch
from transformers import PreTrainedTokenizerFast, BartForConditionalGeneration

# KoBART 모델 로드
tokenizer = PreTrainedTokenizerFast.from_pretrained('digit82/kobart-summarization')
model = BartForConditionalGeneration.from_pretrained('digit82/kobart-summarization')

def summarize_text(text):
    """입력된 텍스트를 KoBART를 사용하여 요약"""
    input_ids = tokenizer.encode(text, return_tensors="pt", truncation=True, max_length=1024)
    summary_ids = model.generate(input_ids, num_beams=4, max_length=512, eos_token_id=1)
    return tokenizer.decode(summary_ids.squeeze().tolist(), skip_special_tokens=True)

def chunk_text(text, chunk_size=700):
    """텍스트를 chunk_size 크기로 나누는 함수"""
    words = text.split()  # 공백 기준으로 단어 단위 분할
    chunks = []
    for i in range(0, len(words), chunk_size):
        chunks.append(" ".join(words[i:i + chunk_size]))
    return chunks

# 문서를 chunk 단위로 나누기
chunks = chunk_text(ex_pdf, chunk_size=700)
print(chunks)

# 각 chunk 요약 수행
summaries = [summarize_text(chunk) for chunk in chunks]

#1차 요약된 결과를 다시 합쳐서 최종 요약
final_summary = summarize_text(" ".join(summaries))

print(final_summary)
```

```python
import PyPDF2
import re
from transformers import PreTrainedTokenizerFast, BartForConditionalGeneration

# 1. PDF 파일에서 텍스트 추출
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in reader.pages:
            text += page.extract_text()
    return text

# 2. 텍스트 클리닝 (불필요한 공백 및 특수문자 제거)
def clean_text(text):
    text = re.sub(r'\n+', '\n', text)  # 여러 줄 바꿈을 하나로
    text = re.sub(r'\s+', ' ', text)  # 여러 공백을 하나로
    text = re.sub(r'[^\x00-\x7F]+', '', text)  # 비 ASCII 문자 제거 (필요시)
    return text.strip()

# 3. 텍스트를 청크로 나누기
def split_text_into_chunks(text, max_length=1024):
    # 텍스트를 문장 단위로 나누고, 각 청크의 길이를 제한
    sentences = text.split('. ')
    chunks = []
    chunk = ""
    for sentence in sentences:
        if len(chunk) + len(sentence) < max_length:
            chunk += sentence + '. '
        else:
            chunks.append(chunk.strip())
            chunk = sentence + '. '
    if chunk:
        chunks.append(chunk.strip())
    return chunks

# 4. koBART로 텍스트 요약하기
def summarize_with_kobart(text_chunks):
    # 모델 및 토크나이저 로드
    tokenizer = PreTrainedTokenizerFast.from_pretrained('digit82/kobart-summarization')
    model = BartForConditionalGeneration.from_pretrained('digit82/kobart-summarization')

    summaries = []
    for chunk in text_chunks:
        inputs = tokenizer(chunk, return_tensors="pt", max_length=1024, truncation=True)
        summary_ids = model.generate(inputs['input_ids'], max_length=150, num_beams=4, early_stopping=True)
        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        summaries.append(summary)
    return summaries

# 5. 요약 결과 합치기
def combine_summaries(summaries):
    return ' '.join(summaries)

# 실행 예시
pdf_path = './example.pdf'  # 처리할 PDF 파일 경로
raw_text = extract_text_from_pdf(pdf_path)
print(raw_text)
cleaned_text = clean_text(raw_text)
print(cleaned_text)
chunks = split_text_into_chunks(cleaned_text)
print(chunks)

summaries = summarize_with_kobart(chunks)
final_summary = combine_summaries(summaries)

print(final_summary)
```
